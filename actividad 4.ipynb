```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cufflinks as cf
from IPython.display import display,HTML 
```

```python
#Leo mi archivo.csv
covid = pd.read_csv("COVID19Mexico (1).csv")
#Filtro por filas
covid= covid.iloc[0:100000,: ]
covid
```

```python
valores_nulos=covid.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
covid=covid.copy()
```

```python
#Sustituir valores nulos con mediana
covid["SECTOR"]=covid["SECTOR"].fillna(round(covid["SECTOR"].median(),1))
covid
```

```python
#Sustituir valores nulos con mediana
covid["ENTIDAD_UM"]=covid["ENTIDAD_UM"].fillna(round(covid["ENTIDAD_UM"].median(),1))
covid
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
covid["SEXO"] =covid["SEXO"].fillna(method="ffill") 
covid
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
covid["MUNICIPIO_RES"] =covid["MUNICIPIO_RES"].fillna(method="ffill") 
covid
```

```python
valores_nulos=covid.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con mediana
covid["EDAD"]=covid["EDAD"].fillna(round(covid["EDAD"].median(),1))
covid
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
covid["EMBARAZO"] =covid["EMBARAZO"].fillna(method="ffill") 
covid
```

```python
covid["PAIS_NACIONALIDAD"] =covid["PAIS_NACIONALIDAD"].fillna(method="ffill") 
covid
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=covid["ORIGEN"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Ouliers
outliers_iqr= covid[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr= covid[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_iqr
```
